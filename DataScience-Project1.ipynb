# Clean Data:

```python
# Convert dates to datetime
disasters['Start Date'] = pd.to_datetime(disasters['Start Date'])
disasters['End Date'] = pd.to_datetime(disasters['End Date'])
disasters['Close Date'] = pd.to_datetime(disasters['Close Date'])
disasters['Declaration Date'] = pd.to_datetime(disasters['Declaration Date'], format='%m/%d/%Y')
```

```python
# Calculate the duration of each disaster
def to_days(array):
  return [(str(i).split()[0]) for i in array]
disasters['Duration (days)'] = to_days(disasters['End Date'] - disasters['Start Date'])
disasters['Duration (days)'].replace({"NaT": None}).astype("float")

# Extract year from Declaration Date
disasters['Year'] = disasters['Declaration Date'].dt.year

# Calculate response time as difference between Start Date and Declaration Date
disasters['Response Time'] = (disasters['Declaration Date'] - disasters['Start Date']).dt.days
```

1. What are the most common types of natural disasters in the US and how long to they tend to last?
The five most common types of natural diasters declared in the US and their durations are:

```python
# Most frequent disasters in the US (1952-2017)
disaster_counts = disasters.groupby('Disaster Type').size().reset_index(name='counts')
disaster_counts = disaster_counts.sort_values('counts', ascending=False)
disasters_hist = alt.Chart(disaster_counts).mark_bar().encode(
    y=alt.Y('Disaster Type', sort='-x'),
    x='counts',
)
disasters_hist.display()
```

```python
# Duration of disaster types in the US from 1952-2017
disaster_durations =  disasters.groupby(['Disaster Type', 'Duration (days)', 'Start Date']).size().reset_index(name='counts').sort_values('counts', ascending=False)

selection = alt.selection_multi(fields=['Disaster Type'], bind='legend')

disasters_scatter = alt.Chart(disaster_durations).mark_circle().encode(
    x='Start Date:T',
    y='Duration (days):Q',
    color='Disaster Type',
    opacity=alt.condition(selection, alt.value(1), alt.value(0.01)),
    tooltip='Duration (days)'
).add_selection(
    selection).interactive()

disasters_scatter.display()
```

2. How have the frequency of different types of disasters changed over time?
Very few disasters were declared in the first decade of our data between 1953 and 1936: between 7 and 23 declarations for any given year during that period. The frequency of recorded emergencies increased, averaging around 300 cases per year until 1993. There was a huge increase in subsequent years, averaging approximately 1,000 disasters per year with the largest frequency occuring in 2005 with 4,736 cases, nearly doubling that of even the second highest year.

Occurences of most disaster types have been rare between 1952-2017. Those of note are listed below:

1.)There was a spike in droughts between 1995-1997 to over 200 declarations and again in the mid-1990s to 150 cases. None have been declared since 2008.

2.)Fires were infrequent until 1994 when yearly spikes averaging around 50 cases occured until 2003. After this, cases spiked in 2005 to nearly 400, then again in 2008 (approx. 220 cases) and 2011 (approx. 130 cases).

3.)Flooding has been a frequent and variable occurence, averaging about 25 cases between 1965-1975, 10 cases in the subsequent decade, and approximately 50 cases since 1996 with lull in the early 2000s.

4.)Hurricane declarations were relatively infrequent until 1996 when they began averaging around 100 cases per year with a spike to over 2,500 cases in 2005 as a result of Hurricane Katrina.

5.)Ice related disasters were not frequently declared until 2000 averaging around 50 cases per year.
6.)Disasters in the "other" category only occurred between 1998 and 2006, rising and falling steadily through the period with a peak in 2003 at 100 cases.

7.)Snow related disasters were infrequent until cases spiked in 1993 to 800 cases, after which smaller peaks have been observed biannually.

8.)Storm disasters have been frequent since 1998 and average 50 cases per year with some years having over 150 cases.

9.)It appears that the terrerosim category of disasters was not added until the 9/11 attack, and no emergencies of this type have been declared since.

10.0Tornado emergencies 1974 with 152 cases with an average of 20 cases per year.

```python
# Sort the dataframe by duration in descending order
disasters_sorted = disasters.sort_values('Duration (days)', ascending=False)

# Select the disaster type with the maximum duration
max_duration_disaster_type = disasters_sorted.iloc[0]['Disaster Type']

# Filter the dataframe to only include rows with the maximum duration disaster type
disasters_filtered = disasters_sorted[disasters_sorted['Disaster Type'] == max_duration_disaster_type]

# Group the filtered dataframe by state and count the number of disasters in each state
disasters_grouped = disasters_filtered.groupby(['State',]).count().reset_index()

# Create a bar chart with state on the x-axis and the count of disasters on the y-axis
chart = alt.Chart(disasters_grouped).mark_bar().encode(
    x='State',
    y='Duration (days)',
    tooltip='Duration (days)'
).interactive()

# Set the chart title
chart = chart.properties(title='Maximum Duration of a Disaster by State')

# Display the chart
chart.display()
```

4. How have of disaster assistance programs historically been allocated to different types of disasters and over time?

Individual Assistance Program:
This program is most frequently allocated to storm and flood disasters. Its use became more frequent after 1965 with the most disasters (300 cases) recieving this assistance in 2005, likely in response to Hurricane Katrina.

Individuals & Households Program:
This program is most frequently allocated to storm and hurricane disasters. Its use began in 2003 and was freuqently allocated to disasters with an average of about 100 cases per year until 2006 when the average usage dropped to approx. 50 cases per year. This followed a spike to over 300 cases in 2005, likely in response to Hurricane Katrina.

Public Assistance Program:
This program is most frequently allocated to storm, flood, and hurricane disasters. Its use became more frequent in 1960. After a decline in the 1980s, this program has been more regularly allocated to disasters in recent years averaging approx. 75 cases per year with a spike to nearly 800 in 2005, likely in response to Hurricane Katrina.

Hazard Mitigation Program:
This program is most frequently allocated to storm and flood disasters. It is used consistently since 1965 and more frequently following a spike to over 450 cases in 1996 and averaging approx. 75 cases per year since.

```python
filtered_df = disasters[disasters['Individual Assistance Program'] == 'Yes']

# Group the filtered data by Disaster Title and count the number of disasters with Individual Assistance Program as Yes
grouped_df = filtered_df.groupby(['Disaster Type'])['Individual Assistance Program'].count().reset_index(name='count')

# Create the bar chart
chart = alt.Chart(grouped_df).mark_bar().encode(
    x=alt.X('count:Q', title='Number of Individual Assistance Programs'),
    y=alt.Y('Disaster Type:N', title='Disaster Type', sort='-x')
).properties(
    title='Disasters Type which received Individual Assistance Program'
)

# Show the chart
chart
```

5. How has the response time for a natural diasters changed over the years?
The time between the start of natural disasters and their official declaration has increased sharply since 1980. Although variable from year to year, the average response time is approx. 25 days.

```python
# Group data by year and calculate mean response time
mean_response_time = disasters.groupby('Year')['Response Time'].mean().reset_index()

# Create Altair chart
chart = alt.Chart(mean_response_time).mark_line().encode(
    x='Year',
    y='Response Time',
    tooltip=['Year', 'Response Time']
).properties(
    width=500,
    height=300,
    title='Mean Response Time by Year'
).interactive()

# Show the chart
chart
```
