#### **About the Data:**
The dataset is available on Kaggle at https://www.kaggle.com/fema/federal-disasters. It contains information on all natural disasters that have been declared by the President of the United States since 1953. The data is spread across multiple tables and includes information such as the type of disaster, location, date, estimated cost of damage, and assistance provided by FEMA. The main table is available as a csv file called "database" and has 14 columns and over 46,000 rows.

##### **Clean Data:**

```python
# Convert dates to datetime
disasters['Start Date'] = pd.to_datetime(disasters['Start Date'])
disasters['End Date'] = pd.to_datetime(disasters['End Date'])
disasters['Close Date'] = pd.to_datetime(disasters['Close Date'])
```

```python
# Calculate the duration of each disaster
disasters['Duration'] = disasters['End Date'] - disasters['Start Date']
```

##### 1.   **What are the most common types of natural disasters in the US?**
> This provides a general overview of the types of natural disasters that occur in the U.S. and can be quantified using the raw data to create a frequency table and/or histogram.

```python
# Most frequent disasters in the US (1952-2017)
disaster_counts = disasters.groupby('Disaster Type').size().reset_index(name='counts')
disaster_counts = disaster_counts.sort_values('counts', ascending=False)
disasters_hist = alt.Chart(disaster_counts).mark_bar().encode(
    y=alt.Y('Disaster Type', sort='-x'),
    x='counts',
)
disasters_hist.display()
```

```python
# Frequency of disasters by State in the US (1952-Present)
state_counts = disasters.groupby('State').size().reset_index(name='counts')
state_counts = state_counts.sort_values('State', ascending=False)
state_hist = alt.Chart(state_counts).mark_bar().encode(
    y=alt.Y('State', sort='-x'),
    x='counts',
)
state_hist.display()
```

```python
# Sort the dataframe by duration in descending order
disasters_sorted = disasters.sort_values('Duration', ascending=False)

# Select the disaster type with the maximum duration
max_duration_disaster_type = disasters_sorted.iloc[0]['Disaster Type']

# Filter the dataframe to only include rows with the maximum duration disaster type
disasters_filtered = disasters_sorted[disasters_sorted['Disaster Type'] == max_duration_disaster_type]

# Group the filtered dataframe by state and count the number of disasters in each state
disasters_grouped = disasters_filtered.groupby(['State']).count().reset_index()

# Create a bar chart with state on the x-axis and the count of disasters on the y-axis
chart = alt.Chart(disasters_grouped).mark_bar().encode(
    x='State',
    y='Duration'
)

# Set the chart title
chart = chart.properties(title='Disaster Type with Maximum Duration by State')

# Display the chart
chart.display()
```
